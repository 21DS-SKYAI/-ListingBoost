# ListingBoost Quickstart
Small demo notebook to generate a caption programmatically.


```python
from ListingBoost.app.utils import ListingContext, load_prompt_template, render_prompt, CaptionGenerator
import pathlib
ctx = ListingContext(bhk='3', area='1800', locality='Worli', city='Mumbai', price='₹12 Cr', amenities='2 Car Parks, Clubhouse', tone='luxury', highlights='Sea-facing, high floor')
tmpl_path = pathlib.Path('ListingBoost/prompts/base_prompt.txt')
tmpl = load_prompt_template(str(tmpl_path))
prompt = render_prompt(tmpl, ctx)
gen = CaptionGenerator()
print(gen.generate(prompt))
```
